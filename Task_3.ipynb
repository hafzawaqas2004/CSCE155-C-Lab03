```python
# Task 1–10 Combined Python Program

import sys

class Student:
    def __init__(self, student_id=None, student_name=None, student_class=None, marks=None):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class
        self.marks = marks

    def display(self):
        print("Student Data:")
        for key, value in self.__dict__.items():
            print(f"{key} : {value}")
        print()


# 1. Instance & namespace
s1 = Student(101, "Hafza", "10th", 85)
print("1. Namespace of instance:", s1.__dict__)
print()

# 2. Function with optional parameters
def student_data(student_id, student_name=None, student_class=None):
    print("2. Student ID:", student_id)
    if student_name:
        print("Student Name:", student_name)
    if student_class:
        print("Class:", student_class)
    print()

student_data(102)
student_data(103, "Fatima", "9th")

# 3. Class info
print("3. Type of Student class:", type(Student))
print("Keys of __dict__:", Student.__dict__.keys())
print("Module:", Student.__module__)
print()

# 4. Empty classes
class Marks:
    pass

s = Student()
m = Marks()

print("4. Is s instance of Student?", isinstance(s, Student))
print("Is m instance of Marks?", isinstance(m, Marks))
print("Is Student subclass of object?", issubclass(Student, object))
print("Is Marks subclass of object?", issubclass(Marks, object))
print()

# 5. Modify attributes
s2 = Student(104, "Fatima", "8th", 75)
print("5. Original:", s2.student_name, s2.marks)
s2.student_name = "Maheen"
s2.marks = 90
print("Updated:", s2.student_name, s2.marks)
print()

# 6. Add & delete attributes
s3 = Student(105, "Hafza")
s3.student_class = "7th"
print("6. Before deletion:", s3.__dict__)
del s3.student_name
print("After deletion:", s3.__dict__)
print()

# 7. Display attributes
print("7. Display attributes:")
s1.display()

# 8. Two instances
student1 = Student(106, "Maheen", "10th", 92)
student2 = Student(107, "Fatima", "9th", 88)

print("8. Student Records:")
student1.display()
student2.display()

# 9. Memory analysis
print("9. Memory size of object:", sys.getsizeof(student1))
print("Memory size of attributes:", sys.getsizeof(student1.__dict__))
print()

# 10. Optimized class using __slots__
class StudentOptimized:
    __slots__ = ['student_id', 'student_name']

    def __init__(self, student_id, student_name):
        self.student_id = student_id
        self.student_name = student_name

opt = StudentOptimized(108, "Maheen")

print("10. Memory with __slots__:", sys.getsizeof(opt))
```

```python
class Trapezoid:
    def __init__(self, a, b, height):
        self.a = a
        self.b = b
        self.height = height

    def area(self):
        return 0.5 * (self.a + self.b) * self.height


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


t = Trapezoid(5, 7, 4)
p = Parallelogram(6, 5)

print("Trapezoid Area:", t.area())
print("Parallelogram Area:", p.area())
```

```python
class Trapezoid:
    def area(self):
        return 0.5 * (5 + 7) * 4

class Parallelogram:
    def area(self):
        return 6 * 5

t = Trapezoid()
p = Parallelogram()

if t.area() > p.area():
    print("Trapezoid has larger area")
else:
    print("Parallelogram has larger area")
```

```python
trapezoids = [(5,7,4), (6,8,5)]
parallelograms = [(5,4), (6,5)]

areas = {}

for i, (a,b,h) in enumerate(trapezoids):
    areas[f"Trapezoid_{i+1}"] = 0.5 * (a+b) * h

for i, (base,h) in enumerate(parallelograms):
    areas[f"Parallelogram_{i+1}"] = base * h

print(areas)
```
