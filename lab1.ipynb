# **Grade Calculator**

```python
marks = int(input("Enter marks (1-100): "))

if marks < 50:
    print("Grade F")
elif marks <= 60:
    print("Grade E")
elif marks <= 70:
    print("Grade D")
elif marks <= 80:
    print("Grade C")
elif marks <= 90:
    print("Grade B")
elif marks <= 100:
    print("Grade A")
else:
    print("Invalid marks")
```

## ***Factorial Calculator***

```python
num = int(input("Enter a number: "))

factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial is:", factorial)
```

# ***Fibonacci Series***


```python
terms = int(input("How many terms? "))

a = 0
b = 1

if terms <= 0:
    print("Enter a positive number")
elif terms == 1:
    print(a)
else:
    print(a, end=", ")
    print(b, end=", ")

    for i in range(3, terms + 1):
        c = a + b
        print(c, end=", ")
        a = b
        b = c
```
